# BeautifulSoup and scraping

Documentation: https://beautiful-soup-4.readthedocs.io/en/latest/


````python
from bs4 import BeautifulSoup
import requests

from PIL import Image
from io import BytesIO

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
````

````python
google_image_text = "cat"

html_page = requests.get(f"https://www.google.com/search?q={google_image_text}&tbm=isch")
soup = BeautifulSoup(html_page.content, "html.parser")

soup.head()
````

````python
print(soup.prettify())
````

````python
soup.title
````

````python
soup.title.get_text()
````

````python
first_div = soup.find("div")
first_div.get_text()
````

````python
first_div.find_next_siblings("div")[1].get_text()
````

# Data Viz
View the most used words with the word cloud

````python
!pip install wordcloud
````

````python
wordcloud = WordCloud().generate(first_div.find_next_siblings("div")[1].get_text())
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
````

````python
# remove stopwords
stopwords = set(STOPWORDS)
stopwords.add("co")
stopwords.add("uk")
stopwords.add("en")
stopwords.add("org")
stopwords.add("app")
stopwords.add("gets")

wordcloud = WordCloud(background_color="white", stopwords=stopwords).generate(first_div.find_next_siblings("div")[1].get_text())
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
````

# Scraping and Computer Vision

````python
images = soup.findAll('img')
len(images)
````

````python
example = images[10]
example
````

````python
example.attrs['src']
````

````python
response = requests.get(example.attrs['src'])
img = Image.open(BytesIO(response.content))

display(img)
````

````python
for image in images:
  url = image.attrs['src']
  if "http" in url:
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))

    display(img)
````

````python
from transformers import pipeline
````

We will automatically organize our images by category

````python
# More models in the model hub.
model_name = "openai/clip-vit-base-patch32"
classifier = pipeline("zero-shot-image-classification", model = model_name)
````

To do that, we will automatically evaluate whether a cat is happy or sad.

But we could also have tested the cat color:

````python
labels_for_classification =  [
                              "happy cat",
                              "sad cat",]

for image in images:
  url = image.attrs['src']
  if "http" in url:
    response = requests.get(image.attrs['src'])
    img = Image.open(BytesIO(response.content))

    display(img)
    print(classifier(img, candidate_labels = labels_for_classification))
````
